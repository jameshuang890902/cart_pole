```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_name = 'RL_data_origin1.npy'
with open(file_name, 'rb') as f:
    rewards = np.load(f)

df = pd.DataFrame(rewards)
avg = df.rolling(20).mean()
```

```python
plt.title('total reward over episode')
plt.xlabel('episode')
plt.ylabel('reward')
plt.plot(rewards,label="reward")
plt.plot(avg,label="20 rolling average reward")
plt.legend()
```

```python
import math
def get_epsilon(i): return max(
    0.01, min(1, 1.0 - math.log10((i+1)/25)))  # epsilon-greedy; 


def get_lr(i): return max(
    0.01, min(0.5, 1.0 - math.log10((i+1)/25)))  # learning rate; 
```

```python
epsilons = np.array([get_epsilon(i) for i in range(200)])
lr = np.array([get_lr(i) for i in range(200)])
```

```python
plt.title('epsilon over episode')
plt.xlabel('episode')
plt.ylabel('epsilon')
# plt.plot(rewards,label="reward")
# plt.plot(avg,label="20 rolling average reward")
plt.plot(lr,label="learning rate")
plt.legend()
```

```python
plt.title('reward over episode')
plt.xlabel('episode')
plt.ylabel('reward')
plt.plot(rewards,label="reward")
plt.plot(avg,label="20 rolling average reward")
plt.plot(epsilons,label="epsilon")
plt.legend()
```
